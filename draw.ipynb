```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
bg = np.zeros((512, 2048, 3), dtype=np.uint8)

# Ensure the polygon coordinates are within the image dimensions
pts = np.array([[300, 200], [900, 200], [900, 400], [200, 400]], np.int32)
pts[:, 0] += 400

# Draw the polygon
cv2.fillPoly(bg, [pts], (255, 255, 255))

# Save the image
cv2.imwrite('input/polygon.png', bg)

# Show the image (convert BGR to RGB for proper display with matplotlib)
plt.imshow(cv2.cvtColor(bg, cv2.COLOR_BGR2RGB))
plt.show()
```
